```python
import os

video_path = "/content/drive/MyDrive/basketball_video.webm"
chunk_dir = "/content/drive/MyDrive/chunks"
os.makedirs(chunk_dir, exist_ok=True)

chunk_duration = 600

# Split video into 10-minute chunks
!ffmpeg -i "{video_path}" -c copy -map 0 -f segment -segment_time {chunk_duration} "{chunk_dir}/chunk%03d.webm"

!ffmpeg -i "{video_path}" -c copy -map 0 -f segment -segment_time {chunk_duration} "{chunk_dir}/chunk%03d.webm"
```

```python
#Transcribe with Whisper
import whisper
import pandas as pd

model = whisper.load_model("medium")

all_segments = []
chunk_files = sorted([f for f in os.listdir(chunk_dir) if f.endswith(".webm")])

for i, file in enumerate(chunk_files):
    file_path = os.path.join(chunk_dir, file)
    print(f"Transcribing {file_path} ...")
    result = model.transcribe(file_path)

    chunk_offset = i * chunk_duration

    for seg in result['segments']:
        all_segments.append([
            seg['start'] + chunk_offset,
            seg['end'] + chunk_offset,
            seg['text']
        ])
```

```python
output_csv = "/content/drive/MyDrive/basketball_transcript_corrected.csv"
df = pd.DataFrame(all_segments, columns=['start_time', 'end_time', 'text'])
df.to_csv(output_csv, index=False)
print(f"Transcript saved to {output_csv}")

# first 5 rows
df.head()
```
